# B2 - AutoLoader to ingest Customer table

```python
source_path = "abfss://externallocation@bricksstorageaccount1.dfs.core.windows.net/source/"


df = (
    spark.readStream.format("cloudFiles")
         .option("cloudFiles.format", "csv")
         .option("cloudFiles.inferColumnTypes", "true")                
         .option("cloudFiles.schemaLocation", "abfss://externallocation@bricksstorageaccount1.dfs.core.windows.net/source/")
         .option("header", "true")
         .option("inferSchema", "true")
         .load(source_path)
)

df.writeStream.format("delta")\
    .option("checkpointLocation", "abfss://externallocation@bricksstorageaccount1.dfs.core.windows.net/checkpoint/")\
    .option("delta.columnMapping.mode", "name")\
    .option("delta.enableChangeDataFeed", "true")\
    .outputMode("append")\
    .trigger(availableNow=True)\
    .table('decisionminds.bronze.customer')
```

```python
%sql
select * from decisionminds.bronze.customer;
```
